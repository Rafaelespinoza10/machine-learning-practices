## Representación gráfica de datos

Si a la función de trazado se le da una matriz de datos, la usará como coordenadas en el eje vertical, y utilizará el índice de cada punto de datos en el array como la coordenada horizontal.

```python
list = [1, 2, 3, 5, 7, 8,11,12,13, 20]
plt.plot(list)
plt.show()
```

También se puede proporcionar dos matrices: una para el eje horizontal, y otra para el eje vertical.

```python

list_num2 = [1, 4, 6, 7, 8, 10, 12, 15, 17, 20]
plt.plot(list, list_num2)
plt.show()
```

Pueden modificarse las logitudes de los ejes para que la figura no se vea tan ajustada

```python
plt.plot([-3, -1, 0, 4, 7], [1, 4, 6, 7, 8])
plt.show()
```

```python
plt.plot([-3, -1, 0, 4, 7], [1, 4, 6, 7, 8])
plt.axis([-4, 8, 0, 10])
plt.show()
```

Se sigue el mismo procedimiento para pintar una función matemática

```python
import numpy as np
```

```python
x = np.linspace(-2, 2, 500)
y = x**2 + 5
plt.plot(x, y)
plt.show()
```

También pude modificarse el estilo de la gráfica para que contenga más información.

```python
plt.plot(x, y)
plt.title("Square function")
plt.xlabel("x")
plt.ylabel("f(x)=x**2 + 5")
plt.grid(True)
plt.legend("square function")
plt.show()
```

Pueden superponerse gráficas y cambiar el estilo de las líneas

```python
import numpy as np
```

```python
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1
plt.plot(x,y,'b--',x, y2, 'r')
plt.grid(True)
plt.show()
```

```python
# Separando en diferentes lineas las funciones
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1
plt.plot(x,y,'b--')
plt.plot(x, y2, 'r')
plt.grid(True)
plt.show()
```

Para poder diferenciar entre ambas funciones siempre es recomendable añadir una leyenda

```python
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1
plt.plot(x,y,'b--', label="x**2")
plt.plot(x, y2, 'r', label="x+1")
plt.legend(loc="best")
plt.title("Mathematic Functions")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```

Tambien puede crearse dos graficas que no se superpongan. Estas graficas se organizan en un grid y se denominan subplots.

```python
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.subplot(1, 2, 1)
plt.plot(x,y,'b--', label="x**2")
plt.title("Mathematic Functions")
plt.legend(loc="best")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(x, y2, 'r', label="x+1")
plt.legend(loc="best")
plt.title("Mathematic Functions")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)

plt.show()
```

Para que las gráficas no queden tan ajustadas, podemos hacer la figura más grande.

```python
plt.figure(figsize=(14,8))
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.subplot(1, 2, 1)
plt.plot(x,y,'b--', label="x**2")
plt.title("Mathematic Functions", fontsize=20)
plt.legend(loc="best")
plt.xlabel("x", fontsize=14)
plt.ylabel("f(x)",fontsize=14)
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(x, y2, 'r', label="x+1")
plt.legend(loc="best")
plt.title("Mathematic Functions", fontsize=20)
plt.xlabel("x", fontsize=14)
plt.ylabel("f(x)", fontsize=14)
plt.grid(True)

plt.show()
```

## Histogramas

```python
data = [1, 1.2,1.2, 1.3, 1.5, 2.3, 2.6, 3.1, 3.3, 3.5, 3.9]

plt.subplot(211)
plt.hist(data, bins=15, rwidth=0.8)
plt.grid(True)
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.subplot(212)
plt.hist(data, bins=5, rwidth=0.8)
plt.grid(True)
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```
